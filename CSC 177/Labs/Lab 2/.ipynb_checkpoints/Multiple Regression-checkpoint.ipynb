```python
#The purpose of this cell is to shuffle the dataframes in the dataset in order to
#randomize the rows to ensure that the data is equally distributed before partitioning
#into training data and test data

import os
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
import matplotlib.pyplot as plt

#So here we will open the file, and then shuffle the Dataframe/rows
fileName = '/Users/diannamelendezgomez/Desktop/admissions_2.csv'
df = pd.read_csv(fileName, na_values=['NA','?'])
df = df.reindex(np.random.permutation(df.index))
df.reset_index(inplace=True, drop=True)
# use inplace=False
df
```

```python
#Now that the rows data has been shuffled around, we can split it 80/20 
numInstances = 500
numTrain = 100   # In our data set, we have 500 rows. 100 rows is 20% of the data 
numTest = numInstances - numTrain

trainingData = df[:numTest]
testData = df[numTest:]

print (trainingData)
print (testData)
```
